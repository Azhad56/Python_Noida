```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
def draw_img(img,img_title="Image"):
    plt.imshow(img,cmap="gray")
    plt.title(img_title+str(img.shape))
    plt.axis('off')
    plt.show()
```

```python
img = cv2.imread('/home/azhad56/Desktop/Python_noida/datasets/cody2.jpg')
```

```python
img = cv2.resize(img,(100,100))
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
draw_img(img,img_title='RGB IMAGE')
gray_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
draw_img(gray_img,img_title='GRAY IMAGE')
```

```python
def convolutionold(img,img_filter):
    W = img.shape[0]
    H = img.shape[1]
    F = img_filter.shape[0]
    output_img = np.zeros((W-F+1,H-F+1))
    for row in range(W-F+1):
        for col in range(H-F+1):
            for i in range(F):
                for j in range(F):
                    output_img[row][col] += img[row+i][col+j]*img_filter[i][j]
                if output_img[row][col] > 255:
                    output_img[row][col] = 255
                elif output_img[row][col] <0:
                    output_img[row][col] = 0
    return output_img
```

```python
blur_filter = np.ones((3,3))/9.0
```

```python
blur_image = convolutionold(gray_img,blur_filter)
```

```python
draw_img(blur_image)
```

```python
edge_filter = np.array([[1,0,-1],
                      [1,0,-1],
                      [1,0,-1]])
```

```python
edge_image = convolutionold(gray_img,edge_filter)
```

```python
draw_img(edge_image)
```

```python
X = np.array([[1,0,2,3],
              [4,6,6,8],
              [3,1,1,0],
              [1,2,2,4]])
def pooling(X,mode="max"):
    W,H = X.shape
    f = 2
    s = 2
    Ho = int((W-f)/s) + 1
    Wo = int((H-f)/s) +1
    pool_output = np.zeros((Wo,Ho))
    for r in range(Wo):
        for c in range(Ho):
            r_start = r*s
            r_end = r_start + f
            c_start = c*s
            c_end = c_start + f
            region = X[r_start:r_end,c_start:c_end]
            if mode=="max":
                pool_output[r][c] = np.max(region)
            elif mode=="average":
                pool_output[r][c] = np.mean(region)
    return pool_output
```

```python
pooling(X)
```
