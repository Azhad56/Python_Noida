```python
class LinearRegCustom:
    def __init__(self,lr=1,max_steps = 10):
        self.lr = lr
        self.max_steps = max_steps
    def fit(self,X,y):
        self.coef_ = np.random.random()
        self.intercept_ = np.random.random()
        for i in range(self.max_steps):
            print(self.coef_,self.intercept_)
            self.gradient_descent(X,y)
    def gradient_descent(self,X,y):
        delta_coef_,delta_intercept_ = self.gradient(X,y)
        self.coef_ += delta_coef_
        self.intercept_ += delta_intercept_
    def gradient(self,X,y):
        yp = self.predict(X)
        d_e2m = np.mean((yp-y)*X)
        d_e2c = np.mean((yp-y))
        return -d_e2m * self.lr, -d_e2c * self.lr
    def predict(self,X):
        return X * self.coef_ + self.intercept_
```

```python
lr = LinearRegCustom(lr=0.01,max_steps=2000)
```

```python
lr.fit(X,y)
```

```python
lr.intercept_,lr.coef_
```

```python
y_pred = lr.predict(X)
```

```python
plt.scatter(X,y)
plt.scatter(X,y_pred)
```
