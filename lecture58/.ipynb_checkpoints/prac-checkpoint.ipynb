```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10,6
from datetime import datetime
```

```python
data = pd.read_csv('air-passengers/AirPassengers.csv')
```

```python
data
```

```python
data['Month'] = pd.to_datetime(data['Month'],infer_datetime_format=True)
```

```python
data
```

```python
IndexData = data.set_index('Month')
```

```python
IndexData
```

```python
plt.xlabel('Date')
plt.ylabel('#Passenger')
plt.plot(IndexData)
```

```python
#Determining rolling statistics
rollmean = IndexData.rolling(window=12).mean()
rollstd = IndexData.rolling(window=12).std()
```

```python
rollmean.iloc[:20,:]
```

```python
IndexData[1:13].mean()
```

```python
#plot rolling mean and statistics
plt.plot(IndexData,color='red',label="original")
plt.plot(rollmean,color='blue',label = "rolling mean")
plt.plot(rollstd,color="green",label="rolling std")
plt.legend(loc="best")
plt.title("rolling mean vs rolling std")
plt.show(block = False)
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
print("Result of dickey-fuller test")
dftest = adfuller(IndexData['#Passengers'],autolag='AIC')
dfotput= pd.Series(dftest[0:4],index = ['The test statistic','pvalue','usedlag','The number of observations used for the ADF'])
for key,value in dftest[4].items():
    dfotput['Critical value (%s) '%key] = value
print(dfotput)
```

#### For stationarity p-value should be less and critical value should be greater than 'the test statistics'

```python
#Estimating trend 
Indexdatalog = np.log(IndexData)
plt.plot(Indexdatalog)
```

```python
movingavg = Indexdatalog.rolling(window = 12).mean()
movingstd = Indexdatalog.rolling(window = 12).std()
```

```python
plt.plot(Indexdatalog,color='red',label="original")
plt.plot(movingavg,color='blue',label = "rolling moving average")
plt.plot(movingstd,color="green",label="rolling moving std")
plt.legend(loc="best")
plt.title("moving avg vs moving std")
plt.show(block = False)
```

```python
Indexdatalog
```

```python
movingavg
```

```python
datalogMinusMovingavg = Indexdatalog - movingavg
datalogMinusMovingavg.dropna(inplace = True)
```

```python
datalogMinusMovingavg.head()
```

```python
def teststationarity(timeseries):
    movingavg = timeseries.rolling(window = 12).mean()
    movingstd = timeseries.rolling(window = 12).std()
    
    plt.plot(timeseries,color='red',label="original")
    plt.plot(movingavg,color='blue',label = "rolling moving average")
    plt.plot(movingstd,color="green",label="rolling moving std")
    plt.legend(loc="best")
    plt.title("moving avg vs moving std")
    plt.show(block = False)
    
    print("Result of dickey-fuller test")
    dftest = adfuller(timeseries['#Passengers'],autolag='AIC')
    dfotput= pd.Series(dftest[0:4],index = ['The test statistic','pvalue','usedlag','The number of observations used for the ADF'])
    for key,value in dftest[4].items():
        dfotput['Critical value (%s) '%key] = value
    print(dfotput)
```

```python
teststationarity(datalogMinusMovingavg)
```

```python
expDecayWeightedAvg = Indexdatalog.ewm(halflife=12,min_periods=0,adjust = True).mean()
```

```python
plt.plot(Indexdatalog)
plt.plot(expDecayWeightedAvg)
```

```python
datalogMinusexpDecayWeightedAvg = Indexdatalog - expDecayWeightedAvg

datalogMinusexpDecayWeightedAvg.dropna(inplace = True)
```

```python
plt.plot(Indexdatalog)
plt.plot(datalogMinusexpDecayWeightedAvg)
```

```python
teststationarity(datalogMinusexpDecayWeightedAvg)
```

```python
datalogdiffshifting = Indexdatalog - Indexdatalog.shift()
```

```python
plt.plot(datalogdiffshifting)
```

```python
datalogdiffshifting.dropna(inplace=True)
teststationarity(datalogdiffshifting)
```
