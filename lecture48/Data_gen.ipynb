```python
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
path = "../datasets/Images/horses/horse-1.jpg"
```

```python
img = image.load_img(path,target_size=(100,100))
```

```python
img
```

```python
arr_img = image.img_to_array(img)
```

```python
norm_img = arr_img/255.0
```

```python
plt.imshow(norm_img)
```

```python
norm_img.shape
```

```python
x = norm_img.reshape(1,100,100,3)
```

```python
data_gen = image.ImageDataGenerator(rotation_range=30,
                                    width_shift_range=0.2,
                                    height_shift_range=0.2,
                                    shear_range=0.2,
                                   horizontal_flip=True,
                                   fill_mode="nearest")
```

```python
i=0
for batch in data_gen.flow(x,batch_size=1):
    plt.figure()
    plt.imshow(image.array_to_img(batch[0]))
    plt.axis('off')
    i+=1
    if i>=10:
        break
```
