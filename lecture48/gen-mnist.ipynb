```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.utils import to_categorical
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
batch = x_train.reshape(-1, 28, 28, 1)
y_train_hot = to_categorical(y_train, num_classes=10)

batch.shape

generator = ImageDataGenerator(rotation_range=10, horizontal_flip=True, shear_range=1.1)

gen = generator.flow(batch, y_train_hot, batch_size=100)
```

```python
x_out, y_out = gen.next()
```
