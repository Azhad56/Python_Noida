```python
import os,shutil
import matplotlib.pyplot as plt
from keras.preprocessing import image
from keras.layers import *
from keras.models import Sequential
from keras import optimizers
from keras.preprocessing.image import ImageDataGenerator
```

```python
sample_path = "Images/cats/cat.1.jpg"
img = image.load_img(sample_path)
```

```python
model = Sequential()
model.add(Convolution2D(32,(3,3),activation='relu',input_shape = (150,150,3)))
model.add(MaxPool2D(2,2))
model.add(Convolution2D(64,(3,3),activation='relu'))
model.add(MaxPool2D(2,2))
model.add(Convolution2D(128,(3,3),activation='relu'))
model.add(MaxPool2D(2,2))
model.add(Convolution2D(128,(3,3),activation='relu'))
model.add(MaxPool2D(2,2))
model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dense(4,activation='softmax'))
model.summary()
```

```python
adam = optimizers.adam(lr=1e-4)
model.compile(optimizer=adam,loss='categorical_crossentropy',metrics=['acc'])
```

```python
img_generator = ImageDataGenerator(rescale=1.0/255)
```

```python
train_gen = img_generator.flow_from_directory("Images/",target_size=(150,150),class_mode='categorical',batch_size=32)
```

```python
val_gen = img_generator.flow_from_directory("val_images/",target_size=(150,150),class_mode='categorical',batch_size=32)
```

```python
hist = model.fit_generator(train_gen,steps_per_epoch=7,epochs=20,validation_data=val_gen,validation_steps=4)
```

```python
h = hist.history
```

```python
plt.style.use('seaborn')
plt.plot(h['loss'],label='Training Loss')
plt.plot(h['val_loss'],label='Validation Loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn')
plt.plot(h['acc'],label='Training acc')
plt.plot(h['val_acc'],label='Validation acc')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
